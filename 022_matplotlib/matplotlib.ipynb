## Nuestro primer gr치fico

En el siguiente ejemplo puedes ver como genera un sencillo gr치fico. Para ello simplemente tenemos que llamar a la funci칩n `plot` con una lista de valores para generar la imagen y `show` para visualizarla.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

Como puedes ver, los valores de la lista se usan para el eje `y` mientras que en el eje `x` se utiliza el 칤ndice de cada valor en la lista. Podemos pasar dos listas a la funci칩n `plot` para indicar los valores en ambos ejes.

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

Puedes visualizar cualquier funci칩n matem치tica generando una lista con los valores de las variables independientes y calculando los valore de las variables dependientes con `Numpy` de la siguiente manera

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

## Customizando un gr치fico

Ahora que conocemos la sintaxis b치sica para crear y visualizar un gr치fico con `Matplotlib` vamos a ver diferente funcionalidad que nos ofrece para tunear nuestro gr치fico y hacerlo m치s bonito. Podemos empezar a침adiendo un t칤tulo y etiquetas a nuestros ejes. En ocasiones tambi칠n es interesante a침adir una cuadr칤cula en el fondo para mejorar la lectura de datos.

```python
plt.plot(x, y)
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

Podemos cambiar el estilo de la l칤nea de manera sencilla pasando diferentes opciones en la funci칩n `plot`, como un string que incluye el formato y color. Puedes encontrar una lista con los diferentes estilos en la [documentaci칩n](https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html).

```python
# l칤nea negra punteada
plt.plot(x, y, '-.k')
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
# diamantes rojos
x2 = np.linspace(-2, 2, 20)
y2 = x2**3

plt.plot(x2, y2, 'dr')
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y = x**3")
plt.grid(True)
plt.show()
```

Podemos a침adir varias curvas en la misma gr치fica simplemente llamando a la funci칩n `plot` varias veces.

```python
plt.plot(x, y, '-.k')
plt.plot(x2, y2, 'dr')
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

A침adiendo una etiqueta a cada curva podemos generar una leyenda, la cual podemos situar en diferentes posiciones.

```python
plt.plot(x, y, '-.k', label="y = x**2")
plt.plot(x2, y2, 'dr', label="y = x**3")
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

Con la funci칩n `axis` podemos delimitar la extensi칩n de los ejes.

```python
plt.plot(x, y, '-.k', label="y = x**2")
plt.plot(x2, y2, 'dr', label="y = x**3")
plt.title("Mi gr치fico")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend(loc='lower right')
plt.axis([-2,2,-4,4])
plt.show()
```

Por 칰ltimo vamos a ver c칩mo cambiar el tama침o del texto para que sea m치s legible. Tambi칠n podemos usar notaci칩n `Latex` para introducir ecuaciones y s칤mbolos en nuestros t칤tulos, ejes y leyendas.

```python
plt.plot(x, y, '-.k', label="$y = x^2$")
plt.plot(x2, y2, 'dr', label="$y = x^3$")
plt.title("Mi gr치fico", fontsize=20)
plt.xlabel("x", fontsize=18)
plt.ylabel("y", fontsize=18)
plt.grid(True)
plt.legend(loc='lower right', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axis([-2,2,-4,4])
plt.show()
```

Ahora que tenemos nuestro gr치fico listo, podemos guardarlo en un archivo simplemente sustituyendo la funci칩n `show` por `savefig`.

```python
plt.plot(x, y, '-.k', label="$y = x^2$")
plt.plot(x2, y2, 'dr', label="$y = x^3$")
plt.title("Mi gr치fico", fontsize=20)
plt.xlabel("x", fontsize=18)
plt.ylabel("y", fontsize=18)
plt.grid(True)
plt.legend(loc='lower right', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axis([-2,2,-4,4])
plt.savefig("mi_grafico.png", transparent=True)
```

## M칰ltiples gr치ficos

Una funcionalidad muy 칰til que `Matplotlib` nos ofrece es la de generar varias gr치ficos en una misma figura. Vamos a ver c칩mo visualizar varias im치genes del dataset MNIST a la vez en una solo figura.

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]

X.shape, y.shape
```

```python
X = X.reshape(-1, 28,28)

X.shape
```

Empezamos definiendo una figura vac칤a con un tama침o determinado con la funci칩n `figure`. Despu칠s, podemos a침adir sub-figuras a la figure con la funci칩n `subplot` indicando su posici칩n en la figura global.

```python
fig = plt.figure(figsize=(3,3))
ax = plt.subplot(1,1,1)
ax.imshow(X[0], cmap="gray")
plt.axis('off')
plt.show()
```

> 丘 La funci칩n `imshow` nos permite visualizar im치genes. Puedes aprender m치s [aqu칤](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html).

## Otros tipos de gr치ficos

Adem치s de la funci칩n `plot`, `Matplotlib` nos ofrece otras funciones para generar gr치ficos de otros tipos. Vamos a ver algunos ejemplos.

### *Scatter*

La funci칩n `scatter` nos permite visualizar un conjunto de puntos.

```python
from numpy.random import rand

x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

Una opci칩n muy interesante en este tipo de gr치ficos es visualizar cada punto con un color o tama침o diferente en funci칩n de otra variable, que puede representar una densidad, temperatura en una localizaci칩n, etc.

```python
for color in ['red', 'green', 'blue']:
    n = 100
    x, y = rand(2, n)
    scale = 500.0 * rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)

plt.show()
```

### Histogramas

Otro tipo de gr치fico muy 칰til es el histograma de barras. Usamos la funci칩n `hist` a la cual podemos indicarle el n칰mero de barras con la opci칩n `bins`.

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.hist(data, bins = 10, rwidth=0.8)
plt.show()
```

## Animaciones

Vamos a terminar viendo c칩mo generar animaciones, una funcionalidad muy 칰til que nos permite generar v칤deos o *gifs* que pueden quedar muy bien en presentaciones. En el siguiente ejemplo veremos como generar una animaci칩n similar a la usada en el primer post del [Perceptr칩n](https://sensioai.com/blog/012_perceptron1) en el que vimos la evoluci칩n de nuestro modelo a medida que lo entrenamos. 

```python
# datos para entrenar el perceptr칩n

np.random.seed(42)

x = np.random.rand(20)
y = 2*x + (np.random.rand(20)-0.5)*0.5

plt.plot(x, y, "b.")
plt.xlabel("$x_1$", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=14)
plt.grid(True)
plt.show()
```

```python
# funciones necesarias para entrenar el perceptr칩n

def gradient(w, x, y): 
    # calculamos la derivada de la funci칩n de p칠rdida
    # con respecto a los par치mteros `w`
    dldw = x*w - y
    dydw = x
    dldw = dldw*dydw
    return np.mean(2*dldw)

def cost(y, y_hat): 
    # calculamos la funci칩n de p칠rdida
    return ((y_hat - y)**2).mean()

def solve(epochs = 29, w = 1.2, lr = 0.2):
    # iteramos un n칰mero determinado de `epochs`
    # por cada epoch, calculamos gradientes y 
    # actualizamos los pesos
    weights = [(w, gradient(w, x, y), cost(x*w, y))]
    for i in range(1,epochs+1):
        dw = gradient(w, x, y)
        w = w - lr*dw
        weights.append((w, dw, cost(x*w, y)))
    return weights
```

Para generar la animaci칩n usamos el objeto `animation` y su funci칩n `FuncAnimation`, a la cual le podemos pasar el n칰mero de *frames* totales en la animaci칩n as칤 como la figura sobre la que queremos aplicar la animaci칩n y la funci칩n con la l칩gica para actualizarla en cada frame. Tambi칠n configuramos `Matplotlib` para visualizar la animaci칩n como un v칤deo en `HTML`.

```python
# animaci칩n

from matplotlib import animation, rc
rc('animation', html='html5')

def update(i):
    xs = np.linspace(0, 1, num=100) 
    (w, dw, cost) = weights[i]
    ax.clear()
    ax.plot(x, y, "b.")
    ax.plot(xs, w*xs, "-k")
    ax.set_xlabel("$x_1$", fontsize=14)
    ax.set_ylabel("$y$", rotation=0, fontsize=14)
    ax.set_title(f"Iteraci칩n: {i}")
    ax.grid(True)
    ax.axis([0,1,0,2])
    return ax
 
weights = solve()
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(1,1,1)
anim = animation.FuncAnimation(fig, update, frames=len(weights), interval=100)
plt.close()

anim
```

Una vez generada la animaci칩n puedes descargarla desde el *widget* creado.
